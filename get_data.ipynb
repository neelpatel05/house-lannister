```python
import json

url = open("GoT_Face_Labelling_Ep7_completed.json","r")
data = url.readlines()
data = json.dumps(data)
data = json.loads(data)
data_new = [json.loads(i) for i in data]
print(data_new[0])
print(len(data_new))
```

```python
from collections import Counter 

count=[]
for i in data_new:
    count.append(i["annotation"]["label"][0])
con  = Counter(count)
for key, value in con.items():
    print("{} : {}".format(key,value))
```

```python
import os
from tqdm import tqdm

os.mkdir("Dataset")
for i in tqdm(data_new):
    if not os.path.exists(os.path.join("Dataset",i["annotation"]["label"][0])):
        os.mkdir("Dataset/"+i["annotation"]["label"][0])
```

```python
fetch_data = {}
url = []
for i in data_new:
    fetch_data[i["annotation"]["label"][0]] = []
for index,value in fetch_data.items():
    for i in data_new:
        if index == i["annotation"]["label"][0]:
            value.append(i["content"])

with open('image_url.json', 'w') as file:
    json.dump(fetch_data, file)
```

```python
import urllib.request

def download_image(name,urls):
    count=0
    for i in tqdm(urls):
        path="Dataset/"+name+"/"+str(count)+".jpg"
        urllib.request.urlretrieve(i, path)
        count=count+1
```

```python
for name, urls in fetch_data.items():
    download_image(name, urls)
```
